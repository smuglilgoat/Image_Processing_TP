```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
from PIL import Image
```

```python
img = cv2.imread('car.jpg')
kernel = np.array([[-1,-2,-1],[0,0,0],[1, 2, 1]])
filteredLin = cv2.filter2D(img,-1,kernel)
plt.imshow(filteredLin)
```

```python
kernel = np.array([[-1,0,1],[-2,0,2],[-1, 0, 1]])
filteredCol = cv2.filter2D(img,-1,kernel)
plt.imshow(filteredCol)
```

```python
linarray = np.asarray(filteredLin)
colarray = np.asarray(filteredCol)
np.square(linarray)
np.square(colarray)
```

```python
arsum = linarray + colarray
np.sqrt(arsum)
reslt = Image.fromarray(arsum, 'RGB')
plt.imshow(reslt, 'gray')
```

```python
artan = np.arctan2(linarray, colarray)
reslt = Image.fromarray(artan, 'RGB')
plt.imshow(reslt, 'gray')
```
